```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

#### reading the csv file

```python
data = pd.read_csv("D:\Rahul\sql practice\datasets\India_GDP_Data.csv")
```

#### heading towards the data visualization.

```python
hist = sns.histplot(data=data, x="GDP_In_Billion_USD")
plt.show
```

#### Boxplot

```python
box = sns.boxplot(x='Year', y='GDP_In_Billion_USD', data = data)
plt.show()
```

#### Scatter plot

```python
scatter = sns.scatterplot(x= 'GDP_In_Billion_USD', y='Year', data=data, size='Per_Capita_in_USD')
plt.show()
```

#### Pair plot

```python
pair = sns.pairplot(data=data, hue = 'GDP_In_Billion_USD')
plt.show()
```
