```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Reading the data

```python
data = pd.read_csv('movies.csv')
```

```python
data.head()
```

#### Checking if there is null value

```python
data.isnull().sum()
```

#### looking for the data types

```python
data.dtypes
```

```python
data.sort_values(by=['gross'], ascending=False)
```

#### Dropping the Duplicates

```python
data['company'].drop_duplicates().sort_values(ascending=False)
```

#### Changing the data type of ("budget", "gross")

```python
data['budget'] = data['budget'].fillna(0)
data['gross'] = data['gross'].fillna(0)
data['budget'] = data['budget'].astype("int")
data['gross'] = data['gross'].astype("int")
```

```python
data.dtypes
```

```python
plt.scatter(x=data['budget'], y=data['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earning')
plt.ylabel('Budget for Film')
plt.show()
```

```python
reg_plot = sns.regplot(x='budget', y='gross', data=data, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
reg_plot
```

#### starting with finding correlation

```python
data.corr(method='pearson') # Correlation matrix
```

#### Making the Correlation matrix

```python
corr_matrix = data.corr(method='pearson')
sns_matrix=sns.heatmap(data=corr_matrix, annot=True)
plt.title('Correlation matrix of numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
sns_matrix
```

#### Looking at the columns of non_numeric

```python
data.head()
```

```python
new_data = data
for col in new_data.columns:
    if (new_data[col].dtype == 'object'):
        new_data[col] = new_data[col].astype('category')
        new_data[col] = new_data[col].cat.codes
new_data
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
corr_matrix = new_data.corr(method='pearson')
sns_matrix=sns.heatmap(data=corr_matrix, annot=True)
plt.title('Correlation matrix of numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
sns_matrix
```

```python
new_data
```

```python
corr_new_data = new_data.corr()
corr_pairs = corr_new_data.unstack()
corr_pairs
```

```python
high_corr = corr_pairs[(corr_pairs) > 0.45]
high_corr = high_corr.sort_values()
high_corr
```
